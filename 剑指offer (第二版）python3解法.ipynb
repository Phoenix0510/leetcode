# 03. 数组中重复的数字

找出数组中重复的数字。


在一个长度为 n 的数组 nums 里的所有数字都在 0～n-1 的范围内。数组中某些数字是重复的，但不知道有几个数字重复了，也不知道每个数字重复了几次。请找出数组中任意一个重复的数字。

示例 1：

输入：
[2, 3, 1, 0, 2, 5, 3]
输出：2 或 3 
 

限制：

2 <= n <= 100000

### 解题思路：
- 方法1：把数组排序，从头到尾扫描排序后的数组，O(nlogn)
- 方法2：利用哈希表，从头到尾扫描数组，O(1)判断哈希表里是否已经包含了该数字，若没有加入哈希表，若有返回该数字。时间复杂度O(n)空间复杂度O(n)
- 方法3：如下，根据数组的特性，从头扫描数组的每个数字，当扫描到下标为i的数字m时，若i==m，扫描下一个数字，若i!=m，并且i下标对应的数字和m下标对应的数字比较，不相等的话交换这两个数字，若这两个数字相等，则返回。时间复杂度：O(n),空间复杂度：O(1)

```python
from typing import List 
 
class Solution:
    def findRepeatNumber(self, nums: List[int]) -> int:
        if len(nums) <= 1:
            return -1
        for i in range(len(nums)):
            if nums[i] == i:
                continue
            if nums[i] == nums[nums[i]]:
                return nums[i]
            nums[nums[i]], nums[i]  = nums[i], nums[nums[i]]
        return -1
    
s = Solution().findRepeatNumber([2,3,1,0,2,5,3])
print(s)
```

### 上面这种方法会修改原数组，下面介绍不修改原数组的值的方法：
- 方法1：创建长度为n+1的辅助数组，然后将数组的值一一复制到辅助数组，数字m应该在辅助数组中的下标m位置，这样很容易发现重复值，空间O(n),时间O(n)
- 方法2：数组长度为n+1时，数组里的数字范围为1~n，找到中位数 (1+n)//2 m，分别统计1~m的值在数组中出现了几次，若超过数字m一定存在重复值，再把1~m
的中位数找到(1+m)//2再去统计在数组中出现的次数，一直重复下去直到找到重复值,否则，另一半m+1~n的区间里一定包含重复的数字。时间O(nlogn)空间O(1)

```python
class Solution:
    def findRepeatNumber(self, nums: List[int]) -> int:
            le = len(nums)
            if le <= 1:
                return -1
            start = 1
            end = le - 1
            while start <= end:
                mid = start + (end - start) // 2
                c = self.countRange(nums, start, mid)
                if end == start:
                    if c > 1:
                        return start
                    else:
                        break
                if c > (mid - start + 1):
                    end = mid
                else:
                    start = mid + 1
            return -1
 
    def countRange(self, nums: List[int], start: int, end: int):
        if not nums:
            return 0
        c = 0
        for i in range(len(nums)):
            if nums[i] >= start and nums[i] <= end:
                c += 1
        return c
                 
        
```

# 04. 二维数组中的查找

在一个 n * m 的二维数组中，每一行都按照从左到右递增的顺序排序，每一列都按照从上到下递增的顺序排序。请完成一个高效的函数，输入这样的一个二维数组和一个整数，判断数组中是否含有该整数。

 

示例:

现有矩阵 matrix 如下：

[
  [1,   4,  7, 11, 15],
  [2,   5,  8, 12, 19],
  [3,   6,  9, 16, 22],
  [10, 13, 14, 17, 24],
  [18, 21, 23, 26, 30]
]
给定 target = 5，返回 true。

给定 target = 20，返回 false。

 

限制：

0 <= n <= 1000

0 <= m <= 1000

 

注意：本题与主站 240 题相同：https://leetcode-cn.com/problems/search-a-2d-matrix-ii/

### 解题思路： 
- 由于数组特性，从右上角出发，若target==右上角的值，直接返回True,若target<右上角的值，那么右上角对应的列下面的值都比target大这一列都不必考虑，
若target值u>右上角的值，右上角所在的行都不必考虑，一直重复这个过程直到找到target，或者遍历完数组没有对应值则返回False

```python
class Solution:
    def findNumberIn2DArray(self, matrix: List[List[int]], target: int) -> bool:
        if not matrix:
            return False

        row = len(matrix)
        col = len(matrix[0])

        m = 0
        n = col -1

        while m < row and n >= 0:
                if target == matrix[m][n]:
                    return True
                elif target < matrix[m][n]:
                    n = n - 1
                else:
                    m = m + 1
        
        return False
```

# 05. 替换空格

请实现一个函数，把字符串 s 中的每个空格替换成"%20"。

 

示例 1：

输入：s = "We are happy."
输出："We%20are%20happy."
 

限制：

0 <= s 的长度 <= 10000

### 解题思路：
- 使用双指针，从后往前复制，若遇到空格，赋值为02%，指针减3，当遇到普通字符，直接复制指针减1即可，直到p1和p2相遇
- 相关题目：有两个排序数组A1和A2，内存在A1末尾有足够的空余空间容纳A2，实现A1和A2合并，并且所有数字是排序的，此题解题思路和替换空格一致，从尾到头比较A1和A2中的数字，并把较大的数字复制到A1中的合适位置。

```python
class Solution:
    def replaceSpace(self, s: str) -> str:
        p1 = len(s) - 1
        res = list(s)
        n = s.count(' ')
        res = res + [0]*n*2
        p2 = len(res) - 1
        
        while p1 != p2:
            if res[p1] != ' ':
                res[p2] = res[p1]
                p2 = p2 - 1
            else:
                res[p2] = '0'
                res[p2 - 1] = '2'
                res[p2 - 2] = '%'
                p2 = p2 - 3
            p1 = p1 - 1

        return ''.join(res)
```

示例 1：

限制：

0 <= 链表长度 <= 10000
### 解题思路：
- 方法1：利用python的list切片功能倒着返回，res[::-1]
- 方法2：利用栈，先读取ListNode中的值到辅助栈中，然后从栈中读取内容
- 方法3：既然想到利用栈，而递归本质上就是一个栈结构，因此可以用递归实现。

```python
#Definition for singly-linked list.
class ListNode:
    def __init__(self, x):
        self.val = x
        self.next = None

class Solution:
    #反转数组
    def reversePrint(self, head: ListNode) -> List[int]:
        res = []
        while head:
            res.append(head.val)
            head = head.next
        return res[::-1]
    
    # 辅助栈
    def reversePrint2(self, head: ListNode) -> List[int]:
        stack = []
        while head:
            stack.append(head.val)
            head = head.next
        res = []
        while stack:
            res.append(stack.pop())
        return res
    
    # 递归
    def reversePrint2(self, head: ListNode) -> List[int]:
        def recur(node):
            if not node:
                return 
            recur(node.next)
            res.append(node.val)
        res = []
        recur(head)
        return res
        
# head = [1,3,2]           
# s6 = Solution().reversePrint(head)
# print(s6)
```

输入某二叉树的前序遍历和中序遍历的结果，请重建该二叉树。假设输入的前序遍历和中序遍历的结果中都不含重复的数字。

 

例如，给出

前序遍历 preorder = [3,9,20,15,7]
中序遍历 inorder = [9,3,15,20,7]
返回如下的二叉树：

    3
   / \
  9  20
    /  \
   15   7
 

限制：

0 <= 节点个数 <= 5000

 

注意：本题与主站 105 题重复：https://leetcode-cn.com/problems/construct-binary-tree-from-preorder-and-inorder-traversal/

### 解题思路：
- 根据前序遍历序列找到根节点，根据中序遍历序列找到左子树和右子树，然后重复这个过程

```python
# Definition for a binary tree node.
class TreeNode:
    def __init__(self, x):
        self.val = x
        self.left = None
        self.right = None

class Solution:
    def buildTree(self, preorder: List[int], inorder: List[int]) -> TreeNode:
        if len(inorder) == 0:
            return None
        root = TreeNode(preorder[0])
        mid = inorder.index(root.val)
        
        root.left = self.buildTree(preorder[1: mid+1], inorder[:mid])
        root.right = self.buildTree(preorder[mid+1:], inorder[mid+1:])
        
        return root
    
    # 这种方法很慢，还是python自身的切片效率更高，代码还简洁
    def buildTree(self, preorder: List[int], inorder: List[int]) -> TreeNode:
        lp = len(preorder)
        li = len(inorder)
        if lp == 0 or li == 0:
            return None
        root = TreeNode(preorder[0])
        lftin = []
        rgtin = []
        lftpre = []
        rgtpre = []
        idx = 0
        for i in range(li):
            if inorder[i] == root.val:
                idx = i
        # print(idx)
        for i in range(idx):
            # print(i)
            lftin.append(inorder[i])
        # print('lftin', lftin)
        for i in range(idx + 1, li):
            rgtin.append(inorder[i])
        # print('rgtin', rgtin)

        for i in range(1, idx + 1):
            lftpre.append(preorder[i])
        # print('lftpre', lftpre)
        for i in range(idx + 1, lp):
            rgtpre.append(preorder[i])
        # print('rgtpre', rgtpre)
        
        if len(lftpre) > 0:
            root.left = self.buildTree(lftpre, lftin)
        if len(rgtpre) > 0:
            root.right = self.buildTree(rgtpre, rgtin)

        return root
        

pre = [3,9,20,15,7]
ino = [9,3,15,20,7]
s7 = Solution().buildTree(pre, ino)
print(s7)
```

#  09. 用两个栈实现队列

用两个栈实现一个队列。队列的声明如下，请实现它的两个函数 appendTail 和 deleteHead ，分别完成在队列尾部插入整数和在队列头部删除整数的功能。(若队列中没有元素，deleteHead 操作返回 -1 )

 

示例 1：

输入：
["CQueue","appendTail","deleteHead","deleteHead"]
[[],[3],[],[]]
输出：[null,null,3,-1]
示例 2：

输入：
["CQueue","deleteHead","appendTail","appendTail","deleteHead","deleteHead"]
[[],[],[5],[2],[],[]]
输出：[null,-1,null,null,5,2]
提示：

1 <= values <= 10000
最多会对 appendTail、deleteHead 进行 10000 次调用

### 解题思路：
- 栈是先进后出，队列是先进先出，可以使用两个栈实现队列，一个栈存放压入序列，一个栈存放弹出序列，压入序列都压进压入栈，弹出序列都从弹出栈弹出，只要弹出栈中有值就可以弹出，当弹出栈为空时，将压入栈的值都弹出到弹出栈。

## 用两个队列实现栈
### 解题思路：
- 当压入数据时，将数据压入队列A尾部，当弹出数据时，将队列A中除最后一个值外全部依次进入队列B，将A中剩的最后一个元素pop出，此时队列A中为空，压入数据将进入到B中，若弹出数据，则将队列B中除最后一个元素外所有元素再依次进入队列A,将最后一个值弹出，依次轮替交换模拟实现栈。

```python
class CQueue:

    def __init__(self):
        self.stack1 = []
        self.stack2 = []
        
    def appendTail(self, value: int) -> None:
        self.stack1.append(value)

    def deleteHead(self) -> int:
        if self.stack2:
            return self.stack2.pop()
        if not self.stack1:
            return -1
        while self.stack1:
            self.stack2.append(self.stack1.pop())
        return self.stack2.pop()


# Your CQueue object will be instantiated and called as such:
# obj = CQueue()
# obj.appendTail(value)
# param_2 = obj.deleteHead()
```

# 10- I. 斐波那契数列

写一个函数，输入 n ，求斐波那契（Fibonacci）数列的第 n 项（即 F(N)）。斐波那契数列的定义如下：

F(0) = 0,   F(1) = 1
F(N) = F(N - 1) + F(N - 2), 其中 N > 1.
斐波那契数列由 0 和 1 开始，之后的斐波那契数就是由之前的两数相加而得出。

答案需要取模 1e9+7（1000000007），如计算初始结果为：1000000008，请返回 1。

 

示例 1：

输入：n = 2
输出：1
示例 2：

输入：n = 5
输出：5
 

提示：

0 <= n <= 100

### 解题思路：
- 1. 典型的递归实现 自顶向下
- 2. 在递归的基础上增加记忆化搜索功能进行记录，避免重复计算大量已经计算过的值 自顶向下
- 3. 动态规划 自底向上

```python
class Solution:
    # 递归方法
    def fib(self, n: int) -> int:
        if n == 0:
            return 0
        if n == 1:
            return 1
        return self.fib(n - 1) + self.fib(n - 2)
    
    # 记忆化搜索
    def fib1(self, n: int) -> int:
        if n == 0:
            return 0
        if n == 1:
            return 1
        res = [-1] * (n + 1)
        if res[n] == -1:
            res[n] = self.fib(n - 1) + self.fib(n - 2)
        return res[n] % 1000000007
    
    def fib2(self, n: int) -> int:
        if n == 0:
            return 0
        pre = 0
        cur = 1
        for i in range(1, n):
            tmp = cur
            cur = pre + cur
            pre = tmp   
        return cur % 1000000007
        
        
s10 = Solution().fib2(8)
print(s10)
```

一只青蛙一次可以跳上1级台阶，也可以跳上2级台阶。求该青蛙跳上一个 n 级的台阶总共有多少种跳法。

示例 1：

示例 3：

输入：n = 0
输出：1
提示：

0 <= n <= 100
注意：本题与主站 70 题相同：https://leetcode-cn.com/problems/climbing-stairs/


### 解题思路：
- 分析可得这个就是斐波那契数列，解题思路如上题

### 青蛙跳台阶变形：青蛙跳台阶的条件若改成可以青蛙一次可以跳1级、2级...n级，此时该青蛙跳上n级台阶一共多少种方法？
- 2^(n-1)

### 相关题目：一个2x1的小矩形横着或竖着去覆盖更大的矩形，8个2x1的小矩形无重复覆盖2x8的大矩形共多少种方法？
- f(8) = f(7) + f(6) 若一个小矩形竖着覆盖大矩形，此时剩下的部分共f(7)种方法，若一个小矩形横着覆盖大矩形，此时剩下的部分共f(6)种方法，所以此题还是斐波那契数列。

```python
class Solution:
    def numWays(self, n: int) -> int:
        if n == 0:
            return 1
        if n == 1:
            return 1
        pre = 1
        cur = 1
        for i in range(1, n):
            tmp = cur
            cur = pre + cur
            pre = tmp
        return cur % 1000000007
```

# 11. 旋转数组的最小数字

把一个数组最开始的若干个元素搬到数组的末尾，我们称之为数组的旋转。输入一个递增排序的数组的一个旋转，输出旋转数组的最小元素。例如，数组 [3,4,5,1,2] 为 [1,2,3,4,5] 的一个旋转，该数组的最小值为1。  

示例 1：

输入：[3,4,5,1,2]
输出：1
示例 2：

输入：[2,2,2,0,1]
输出：0
注意：本题与主站 154 题相同：https://leetcode-cn.com/problems/find-minimum-in-rotated-sorted-array-ii/

### 解题思路：
- 1. 从头到尾遍历数组找到最小值，时间复杂度O(n)
- 2. 根据数组特性，部分有序，可以使用二分查找的思路来实现，设置两个指针，第一个指针指向数组第一个元素，第二个指针指向数组最后一个元素，根据题目中的旋转规则，第一个元素应该是大于或等于最后一个元素的(除非选择0个元素)，接着找到数组中间元素，如果该元素位于前面的递增序列，那么它应该大于或者等于第一个指针指向的元素，此时数组中的最小值应该位于该中间元素的后面，可以把第一个指针指向该中间元素，反之，则把第二个指针指向该中间元素，这样迭代下去，第一个指针总是指向前面递增序列，第二个指针总是指向后面的递增序列，最终第一个指针指向前面递增序列的最后一个元素，第二个指针指向后面子数组的第一个元素，返回它指向的元素即可。时间复杂度：O(logn)

```python
class Solution:
    def minArray(self, numbers: List[int]) -> int:
        l, r = 0, len(numbers) - 1
        while l < r:
            mid = l + (r - l) // 2
            if numbers[mid] > numbers[r]:
                l = mid + 1
            elif numbers[mid] < numbers[r]:
                r = mid
            else:
                r -= 1
        return numbers[l] # return numbers[r]也一样
```

给你一根长度为 n 的绳子，请把绳子剪成整数长度的 m 段（m、n都是整数，n>1并且m>1），每段绳子的长度记为 k[0],k[1]...k[m-1] 。请问 k[0]*k[1]*...*k[m-1] 可能的最大乘积是多少？例如，当绳子的长度是8时，我们把它剪成长度分别为2、3、3的三段，此时得到的最大乘积是18。

示例 1：

输入: 2
输出: 1
解释: 2 = 1 + 1, 1 × 1 = 1
示例 2:

输入: 10
输出: 36
解释: 10 = 3 + 3 + 4, 3 × 3 × 4 = 36
提示：

2 <= n <= 58
注意：本题与主站 343 题相同：https://leetcode-cn.com/problems/integer-break/


```python
class Solution:
    def cuttingRope(self, n: int) -> int:
        if n <= 1:
            return 0
        if n == 2:
            return 1
        if n == 3:
            return 2
        dp = [0] * (n + 1)
        dp[0] = 0
        dp[1] = 1
        dp[2] = 2
        dp[3] = 3
        for i in range(4, n + 1):
            mx = 0
            for j in range(1, i // 2 + 1):
                tmp = dp[j] * dp[i - j]
                if mx < tmp:
                    mx = tmp
                dp[i] = mx
        return dp[-1]
```

```python
# Definition for singly-linked list.
# class ListNode:
#     def __init__(self, x):
#         self.val = x
#         self.next = None
class Solution:
    def deleteNode(self, head: ListNode, val: int) -> ListNode:
        if head is None:
            return None
        if head.val == val:
            return head.next
        pre, cur = head, head.next
        while cur and cur.val != val:
            pre, cur = cur, cur.next
        if cur:
            pre.next = cur.next
        return head
```

```python
"""
# Definition for a Node.
class Node:
    def __init__(self, x: int, next: 'Node' = None, random: 'Node' = None):
        self.val = int(x)
        self.next = next
        self.random = random
"""
class Solution:
    def copyRandomList(self, head: 'Node') -> 'Node':
        # return copy.deepcopy(head)
        if not head:
            return head
        p = head
        while p:
            node = Node(p.val)
            node.next = p.next
            p.next = node
            p = node.next
        p = head
        while p:
            if p.random:
                p.next.random = p.random.next
            p = p.next.next
        pre = head
        new = head.next
        res = head.next
        while pre:
            pre.next = pre.next.next
            if new.next:
                new.next = new.next.next
            pre = pre.next
            new = new.next
        return res    
```

输入一棵二叉搜索树，将该二叉搜索树转换成一个排序的循环双向链表。要求不能创建任何新的节点，只能调整树中节点指针的指向。

 

为了让您更好地理解问题，以下面的二叉搜索树为例：

 



 

我们希望将这个二叉搜索树转化为双向循环链表。链表中的每个节点都有一个前驱和后继指针。对于双向循环链表，第一个节点的前驱是最后一个节点，最后一个节点的后继是第一个节点。

下图展示了上面的二叉搜索树转化成的链表。“head” 表示指向链表中有最小元素的节点。

 



 

特别地，我们希望可以就地完成转换操作。当转化完成以后，树中节点的左指针需要指向前驱，树中节点的右指针需要指向后继。还需要返回链表中的第一个节点的指针。

 

注意：本题与主站 426 题相同：https://leetcode-cn.com/problems/convert-binary-search-tree-to-sorted-doubly-linked-list/

注意：此题对比原题有改动。


```python
"""
# Definition for a Node.
class Node:
    def __init__(self, val, left=None, right=None):
        self.val = val
        self.left = left
        self.right = right
"""
class Solution:
    def treeToDoublyListCore(self, root: 'Node') -> ('Node', 'Node'):
        """
        :param root: 树的根节点
        :return: 双向链表的 头节点 和 尾节点
        """
        # 根为空，那么 对应的双向链表的 头节点 和 尾节点 也为空
        if root is None:
            return None, None

        # 左子树 对应的 双向链表的头节点和尾节点
        left_head, left_tail = self.treeToDoublyListCore(root.left)
        # 右子树 对应的 双向链表的头节点和尾节点
        right_head, right_tail = self.treeToDoublyListCore(root.right)

        # 根的 左节点 与 左子树的尾节点 互相连接
        # 根的 右节点 与 右子树的头节点 互相连接
        root.left, root.right = left_tail, right_head
        if left_tail:
            left_tail.right = root
        if right_head:
            right_head.left = root

        # 左子树的头节点 如果存在则作为当前 双向链表的头节点，否则使用 根节点。尾节点同理。
        return left_head if left_head else root, right_tail if right_tail else root

    def treeToDoublyList(self, root: 'Node') -> 'Node' or None:
        """
        递归遍历
        时间复杂度：O(N)，所有节点遍历一次。
        空间复杂度：O(N)，当二叉搜索树退化为链表时，树的深度为 N.
        """
        head, tail = self.treeToDoublyListCore(root)
        # 改造成循环双向链表
        if head and tail:
            head.left, tail.right = tail, head
        return head
```

#  42. 连续子数组的最大和

示例1:

输入: nums = [-2,1,-3,4,-1,2,1,-5,4]
输出: 6
解释: 连续子数组 [4,-1,2,1] 的和最大，为 6。
 

提示：

1 <= arr.length <= 10^5
-100 <= arr[i] <= 100
注意：本题与主站 53 题相同：https://leetcode-cn.com/problems/maximum-subarray/


```python
class Solution:
    def maxSubArray(self, nums: List[int]) -> int:
        if not nums:
            return
        dp = [0] * len(nums)
        dp[0] = nums[0]
        mx = nums[0]
        for i in range(1, len(nums)):   
            dp[i] = max(dp[i - 1] + nums[i], nums[i])
            if mx < dp[i]:
                mx = dp[i]
        return mx
```

提示：

0 < grid.length <= 200
0 < grid[0].length <= 200


```python
class Solution:
    def maxValue(self, grid: List[List[int]]) -> int:
        if not grid:
            return None
        row = len(grid)
        col = len(grid[0])
        for i in range(1, col):
            grid[0][i] += grid[0][i - 1]
            # print(dp[0][i])
        for i in range(1, row):
            grid[i][0] += grid[i - 1][0]
            # print(dp[i][0])
        for i in range(1, row):
            for j in range(1, col):
                grid[i][j] += max(grid[i - 1][j], grid[i][j - 1])
        return grid[-1][-1]
```

我们把只包含质因子 2、3 和 5 的数称作丑数（Ugly Number）。求按从小到大的顺序的第 n 个丑数。

 

示例:

输入: n = 10
输出: 12
解释: 1, 2, 3, 4, 5, 6, 8, 9, 10, 12 是前 10 个丑数。
说明:  

1 是丑数。
n 不超过1690。
注意：本题与主站 264 题相同：https://leetcode-cn.com/problems/ugly-number-ii/


```python
class Solution:
    def nthUglyNumber(self, n: int) -> int:
        if n < 1:
            return 0
        u = [1]
        i2, i3, i5 = 0, 0, 0
        for i in range(1, n):
            u2, u3, u5 = u[i2] * 2, u[i3] * 3, u[i5] * 5
            mi = min(u2, u3, u5)
            u.append(mi)
            if mi == u2:
                i2 += 1
            if mi == u3:
                i3 += 1
            if mi == u5:
                i5 += 1
        return u[-1]
```

```python
# Definition for singly-linked list.
# class ListNode:
#     def __init__(self, x):
#         self.val = x
#         self.next = None

class Solution:
    def getIntersectionNode(self, headA: ListNode, headB: ListNode) -> ListNode:
        if not headA or not headB:
            return None
        p1, p2 = headA, headB
        c1, c2 = 0, 0
        while p1:
            c1 += 1
            p1 = p1.next
        while p2:
            c2 += 1 
            p2 = p2.next
        p1, p2 = headA, headB
        if c1 > c2:
            for i in range(c1 - c2):
                p1 = p1.next
        elif c1 < c2:
            for i in range(c2 - c1):
                p2 = p2.next
        # print(p1.val, p2.val)
        while p1 != p2:
            p1 = p1.next
            p2 = p2.next
        return p1
```

示例 1:

输出: 2
示例 2:

输入: nums = [5,7,7,8,8,10], target = 6
输出: 0
 

限制：

0 <= 数组长度 <= 50000

 

注意：本题与主站 34 题相同（仅返回值不同）：https://leetcode-cn.com/problems/find-first-and-last-position-of-element-in-sorted-array/


```python
class Solution:
    def search(self, nums: List[int], target: int) -> int:
        if not nums:
            return 0
        fs = self.getFirstK(nums, target)
        ls = self.getLastK(nums, target)
        return ls - fs + 1

    def getFirstK(self, nums: List[int], k: int) -> int:
        l, r = 0, len(nums) - 1
        
        while l <= r  :
            mid = l + (r - l) // 2
            if nums[mid] < k:
                l = mid + 1  
            elif nums[mid] >= k:
                r = mid - 1
        return l
            
    def getLastK(self, nums: List[int], k: int) -> int:
        l, r = 0, len(nums) - 1
        while l <= r:
            mid = l + (r - l) // 2
            if nums[mid] <= k:
                l = mid + 1
            else:
                r = mid - 1
        return r
```

例如，0、1、2、3、4这5个数字组成一个圆圈，从数字0开始每次删除第3个数字，则删除的前4个数字依次是2、0、4、1，因此最后剩下的数字是3。

示例 2：

输入: n = 10, m = 17
输出: 2
 

限制：

1 <= n <= 10^5
1 <= m <= 10^6


```python
class Solution:
    def lastRemaining(self, n: int, m: int) -> int:
        # if n < 1 or m < 1:
        #     return -1
        # ls = 0
        # for i in range(2, n + 1):
        #     ls = ( ls + m ) % i
        # return ls
        l = [i for i in range(n)]
        ls = 0
        while len(l) > 1:
            idx = (ls + m - 1) % len(l)
            l.pop(idx)
            ls = idx
        return l[0]
```

```python
class Solution:
    def add(self, a: int, b: int) -> int:
        x = 0xffffffff
        a, b = a & x, b & x
        while b:
            a, b = (a ^ b), (a & b) << 1 & x
        return a if a <= 0x7fffffff else ~(a ^ x)
        # result = [a,b]
        # return sum(result)
```

示例:

输入: [1,2,3,4,5]
输出: [120,60,40,30,24]
 

提示：

所有元素乘积之和不会溢出 32 位整数
a.length <= 100000


```python
class Solution:
    def constructArr(self, a: List[int]) -> List[int]:
        if not a:
            return []
        le = len(a)
        b, tmp = [1] * le, 1
        for i in range(1, le):
            b[i] = b[i - 1] * a[i - 1]
        for i in range(le - 2, -1, -1):
            # print(i)
            tmp *= a[i + 1]
            # print(tmp)
            b[i] *= tmp
        return b
```
